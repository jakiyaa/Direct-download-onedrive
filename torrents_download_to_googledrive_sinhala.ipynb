# 3.දැන් ඉතිං<font color='orange'> ටොරෙනාට් එක ඩවුන්ලෝඩ් කරහන් හුත්තෝ  <img src='https://libtorrent.org/img/logo-color-text.png' height="45" alt="RcloneLab"/>


රවුම වගේ එක ඔබලා <img src='https://www.pngfind.com/pngs/b/20-200711_youtube-button-png.png' height="40" /> manget link/torrent link එක පේස්ට් කරලා enter ඔබහන් හුත්තෝ

```python
link = input ("ටොරෙනාට්ලි ලින්ක් දියන් හුත්තෝ") 

import libtorrent as lt
import time
import datetime

ses = lt.session()
ses.listen_on(6881, 6891)
params = {
    'save_path': './drive/MyDrive/Torrents/',  #ඔතෙන්ට ඕන රෙද්දක්  දාගනින් ඕන්නම් changE download directory
    'storage_mode': lt.storage_mode_t(2)}

print(link)

handle = lt.add_magnet_uri(ses, link, params)
ses.start_dht()

begin = time.time()
print(datetime.datetime.now())

print ('Downloading Metadata...')
while (not handle.has_metadata()):
    time.sleep(1)
print ('Got Metadata, Starting Torrent Download...')

print("Starting", handle.name())

while (handle.status().state != lt.torrent_status.seeding):
    s = handle.status()
    state_str = ['queued', 'checking', 'downloading metadata', \
            'downloading', 'finished', 'seeding', 'allocating']
    print ('%.2f%% complete (down: %.1f kb/s up: %.1f kB/s peers: %d) %s ' % \
            (s.progress * 100, s.download_rate / 1000, s.upload_rate / 1000, \
            s.num_peers, state_str[s.state]))
    time.sleep(5)

end = time.time()
print(handle.name(), "දැන් හරි හුත්තෝ ඩවුන්ලෝඩ් වුනා එහනම් කොල්ලෝ අපි ගියා")  

print("Elapsed Time: ",int((end-begin)//60),"min :", int((end-begin)%60), "sec")

print(datetime.datetime.now())
```
